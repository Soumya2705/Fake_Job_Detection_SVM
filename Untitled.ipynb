## Data cleaning

```python
df.isnull().sum()
```

```python
df=df.drop(['job_id'],axis=1)
```

```python
df=df.drop(['department','salary_range'],axis=1)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
text_cols = ['company_profile', 'description', 'requirements', 'benefits']

for col in text_cols:
    df[col] = df[col].fillna('')
```

```python
df.info()
```

```python
cat_cols = ['employment_type', 'required_experience', 
            'required_education', 'industry', 'function']

for col in cat_cols:
    df[col] = df[col].fillna('Unknown')
```

```python
df['location'] = df['location'].fillna('Unknown')
```

## Feature Engineering

```python
df['text'] = df['title'] + " " + df['description'] + " " + df['requirements']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000)
X_text = tfidf.fit_transform(df['text'])
```

```python
X_cat = pd.get_dummies(df[['employment_type', 'required_experience']])
```

```python
from scipy.sparse import hstack

X = hstack([X_text, X_cat])
y = df['fraudulent']
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Train SVM Model

```python
from sklearn.svm import LinearSVC

model = LinearSVC()
model.fit(X_train, y_train)
```

## Evaluation

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

# Class weight balance

```python
from sklearn.svm import LinearSVC

model = LinearSVC(class_weight='balanced')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

# Undersampling


```python
# Separate classes
df_0 = df[df['fraudulent'] == 0]
df_1 = df[df['fraudulent'] == 1]

# Undersample class 0
df_0_sampled = df_0.sample(n=1000, random_state=42)

# Combine back
df_balanced = pd.concat([df_0_sampled, df_1])

# Shuffle
df_balanced = df_balanced.sample(frac=1, random_state=42)
```
